```python
import os
import os.path as path
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import cv2
import math
import imageio
import glob
from skimage import io
import albumentations as A
import json
import random
from numpy import savez_compressed
```

```python
Augmented_json_file = "../Dataset/finalDataset/allImages/AugImages/augmentedJSON.json"
ImageFolderPath = "../Dataset/finalDataset/allImages/AugImages/"

with open(Augmented_json_file) as f:
    imgs_anns = json.load(f)

classNames = []
listOfImages=[]
for Count,val in enumerate(imgs_anns):
    fileName = val["image"]
    ImgId = val["id"]
    #print(fileName)
    for CountLabels,valLabels in enumerate(val["label"]):
        #print(valLabels["original_width"])
        width = valLabels["original_width"]
        height = valLabels["original_height"]
        
        #print(valLabels["polygonlabels"][0])
        if len(classNames) == 0:
            classNames.append(valLabels["polygonlabels"][0])
        if valLabels["polygonlabels"][0] in classNames:
            continue
        else:
            #print(valLabels["polygonlabels"][0])
            classNames.append(valLabels["polygonlabels"][0])
            
    objImgDesc = {
        "width":width,
        "height":height,
        "id":ImgId,
        "file_name":fileName
    }
    listOfImages.append(objImgDesc)
    #print(classNames)
    categories = []
    for i in range(0,len(classNames)):
        objCat = {
            "supercategory": "None",
            "id":i,
            "name":classNames[i]
        }
        categories.append(objCat)
listOfCategories=categories

#print(len(listOfImages))

#print(listOfCategories)
listOfAnnotations = []
cocoJSON = []        
for Count,val in enumerate(imgs_anns):
    getcount = [-1]*len(classNames)
    ImgId = val["id"]
    for CountLabels,valLabels in enumerate(val["label"]):
        for ind in range(0,len(getcount)):
            if valLabels["polygonlabels"][0] == classNames[ind]:
                getcount[ind] += 1
                catID = ind
        getPoints = np.array(valLabels["points"])
        getPoints[:,0] = getPoints[:,0] * width/100
        getPoints[:,1] = getPoints[:,1] * height/100
        poly = (getPoints.flatten()).tolist()
        obj = {
            "id":CountLabels,
            "image_id":ImgId,
            "category_id":catID,
            "segmentation": [poly],
            "bbox": [np.min(getPoints[:,0]), np.min(getPoints[:,1]), np.max(getPoints[:,0])-np.min(getPoints[:,0]), np.max(getPoints[:,1])-np.min(getPoints[:,1])],
            "ignore":0,
            "iscrowd":0,
            "area":((np.max(getPoints[:,0])-np.min(getPoints[:,0]))*np.max(getPoints[:,1])-np.min(getPoints[:,1]))
          }
        listOfAnnotations.append(obj)
#print(listOfAnnotations[1])
AnImage = {
    "images":listOfImages,
    "categories":listOfCategories,
    "annotations":listOfAnnotations
}
#cocoJSON.append(AnImage)

with open(ImageFolderPath + "cocoJSON.json", "w") as outfile:
    json.dump(AnImage, outfile)
    
#print("Coco JSON Saved!")
```
